```python
import pandas as pf
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import keras_ocr
import cv2
from tensorflow import keras
from tensorflow.keras import layers
```

```python
folder_path = '/content/drive/MyDrive/Reciepts/images'

files = os.listdir(folder_path)

images = []

for file_name in files:
    image_path = os.path.join(folder_path, file_name)
    image = cv2.imread(image_path)
    images.append(image)
```

```python
num_images = len(images)
num_cols = 4
num_rows = (num_images + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))

for i, ax in enumerate(axes.flat):
    if i < num_images:
        ax.imshow(images[i])
        ax.axis('off')
    else:
        ax.axis('off')

plt.show()
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
img_sample = img_sample = [images[i] for i in [1, 2, 3, 4]]
```

```python
prediction = pipeline.recognize(img_sample)
```

```python
fig,axs = plt.subplots(nrows=len(img_sample), figsize=(20,20))
for ax, img, predictions in zip(axs,img_sample,prediction):
  keras_ocr.tools.drawAnnotations(image=img, predictions=predictions,ax=ax)
```
